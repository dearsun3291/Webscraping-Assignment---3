## 1. Write a python program which searches all the product under a particular product from www.amazon.in. The product name to be searched will be taken as input from user. For e.g. If user input is ‘guitar’. Then search for guitars.

```python
def amazon_download(url):
    driver = webdriver.Chrome(r"C:\Users\clear\Desktop\webdriver\chromedriver_win32\chromedriver.exe")
    driver.get(url)
    Mobile = input("Enter product to be searched: ")
    driver.find_element_by_id("twotabsearchtextbox").send_keys(Mobile)
    driver.find_element_by_id("nav-search-submit-button").click()
    lists = []
    for i in range(0,3):
        links = driver.find_elements_by_xpath("//a[@class ='a-link-normal a-text-normal']")
        for i in links:
            lists.append(i.get_attribute('href'))
        next_btn = driver.find_element_by_xpath("//ul[@class='a-pagination']/li[7]/a")
        if next_btn.text == 'Next':
            next_btn.click()
            time.sleep(3)

    names = []
    price = []
    brand = []
    rating = []
    no_of_ratings = []
    return_exchange = []
    delivery = []
    avilability =[]
    other_details =[]
    for i in lists[0:100]:
        driver.get(i)
        delay = 10
        WebDriverWait(driver,delay).until(expected_conditions.presence_of_element_located((By.XPATH,"//span[@class ='a-size-large product-title-word-break']")))
        name = driver.find_element_by_xpath("//span[@class ='a-size-large product-title-word-break']").text
        names.append(name)

        try:
            prices = driver.find_element_by_xpath("//span[@class='a-size-medium a-color-price priceBlockBuyingPriceString']").text
            price.append(prices)
            
        except NoSuchElementException:
            price.append('-')
    
        try:
            brand_name = driver.find_element_by_xpath("//table[@class='a-normal a-spacing-micro']/tbody/tr[3]/td[2]").text
            brand.append(brand_name)
        except NoSuchElementException:
            brand.append('-')
        
        try:
            ratings = driver.find_element_by_xpath("//span[@class='a-size-base a-nowrap']/span").text
            rating.append(ratings)
        except NoSuchElementException:
            rating.append('-')
            
        try:
            no_of_rating = driver.find_element_by_xpath("//span[@class='a-size-base a-color-secondary']").text
            no_of_ratings.append(no_of_rating)
        except NoSuchElementException:
            no_of_ratings.append('-')
            
        try:
            return_policy = driver.find_element_by_xpath("//div[@class='a-column a-span3 a-text-center icon-container icon-grid-width celwidget']/span/div[2]/a").text
            return_exchange.append(return_policy)
        except NoSuchElementException:
            return_exchange.append('-')
            
        try:
            exp_delivery = driver.find_element_by_xpath("//div[@class='a-section a-spacing-mini']/b").text
            delivery.append(exp_delivery)
        except NoSuchElementException:
            delivery.append('-')
            
        try:
            avl = driver.find_element_by_xpath("//div[@id='availability']/span").text
            avilability.append(avl)
        except NoSuchElementException:
            avilability.append('-')
            
        try:
            others = driver.find_element_by_xpath("//ul[@class='a-unordered-list a-vertical a-spacing-mini']").text
            other_details.append(others)
        except NoSuchElementException:
            other_details.append('-')
    
    
    driver.close()
    df = pd.DataFrame({})
    df['Brand']=brand
    df['product_Details']=names
    df['price']=price
    df['rating'] = rating
    df['no_of_ratings'] = no_of_ratings
    df['return_exchange'] = return_exchange
    df['avilability']=avilability
    df['delivery'] = delivery
    df['other_details'] = other_details
    df['Urls']= lists[0:100]
    return(df)
```

## 3. Write a python program to access the search bar and search button on images.google.com and scrape 100 images each for keywords ‘fruits’, ‘cars’ and ‘Machine Learning’.


```python
def image_download(url):
    driver = webdriver.Chrome(r"C:\Users\clear\Desktop\webdriver\chromedriver_win32\chromedriver.exe")
    driver.get(url)
    search = input('Item to search: ')
    driver.find_element_by_xpath("//input[@class ='gLFyf gsfi']").send_keys(search)
    driver.find_element_by_xpath("//span[@class ='z1asCe MZy1Rb']").click()
    img_urls = driver.find_elements_by_xpath("//img[@class = 'rg_i Q4LuWd']")
    img_url = []
    for i in  range(0,100):
        img=img_urls[i]
        try:
            img.click()
            time.sleep(2)
            actual_images = driver.find_elements_by_xpath("//a[@class='eHAdSb']/img[1][@class='n3VNCb']")
            for actual_image in actual_images:
                if actual_image.get_attribute('src') and 'https' in actual_image.get_attribute('src'):
                    img_url.append(actual_image.get_attribute('src'))
        except ElementClickInterceptedException or ElementNotInteractableException as err:
            print(err)

    location = input("Enter Location to save files: ")
    os.chdir(location)
    baseDir=os.getcwd()
    import requests
    for i, url in enumerate(img_url):
        file_name = f"{i:100}.jpg"    
        try:
            image_content = requests.get(url).content
    
        except Exception as e:
            print(f"ERROR - COULD NOT DOWNLOAD {url} - {e}")

        try:
            image_file = io.BytesIO(image_content)
            image = Image.open(image_file).convert('RGB')
        
            file_path = os.path.join(baseDir, file_name)
        
            with open(file_path, 'wb') as f:
                image.save(f, "JPEG", quality=85)
            print(f"SAVED - {url} - AT: {file_path}")

        except Exception as e:
            print(f"ERROR - COULD NOT SAVE {url} - {e}")
```

## 7. Write a program to scrap all the available details of top 10 gaming laptops from digit.in.

```python
def laptops_game(url):
    driver = webdriver.Chrome(r"C:\Users\clear\Desktop\webdriver\chromedriver_win32\chromedriver.exe")
    driver.get(url)
    time.sleep(3)
    soup = BeautifulSoup(driver.page_source,'lxml')
    tables = soup.find_all('table')
    dfs = (pd.read_html(str(tables)))
    df0 = pd.DataFrame(dfs[0])
    lst = [1,2,4,6,8,10,12,14,15,17]
    dn = pd.DataFrame()
    for i in lst:
        df1 = pd.DataFrame(dfs[i])
        df1= df1.drop('SPECIFICATION.1', axis =1)
        df1 = df1.T
        df1 = df1.rename(columns=df1.iloc[0]).drop(df1.index[0])
        dn = dn.append(df1, ignore_index = False)
        
    dn.reset_index(drop=True, inplace=True)
    df = pd.concat([df0,dn], axis =1,ignore_index=False )
    driver.close()
    return(df)
```

```python
laptops_game(url= 'https://www.digit.in/top-products/best-gaming-laptops-40.html')
```
